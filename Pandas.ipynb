```python
## find the avg price of car in each seller type
```

```python
df4.groupby('seller_type').agg({'selling_price':['min']})
```

```python
## find the avg,max,min, std for price of car in each seller type
df4.groupby('seller_type').agg({'selling_price':['mean','min','max','std']})
```
